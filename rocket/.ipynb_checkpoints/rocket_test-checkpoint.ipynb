```python
import requests
import json
import pandas as pd
url = "https://launchlibrary.net/1.4/launch/Falcon"
rock = requests.get(url).json()
print(json.dumps(rock, indent = 4, sort_keys = True))
```

```python
print(type(rock["launches"][5]['failreason']))
print(len(rock["launches"]))
fail = 0
for f in range(len(rock["launches"])):
    if str(rock["launches"][f]['failreason']) != "None":
        fail = fail + 1
        
print("fails", fail)
print(rock["launches"][5]['name'])
print(rock['launches'][2]['windowend'])
print(rock['launches'][3]['lsp']['name'])
print(type(rock['launches'][2]['windowend']))
```

```python
name = []
date = []
o_name = []
sf = []
reason_f = []
lat = []
lng = []

for f in range(len(rock['launches'])):
    name.append(str(rock["launches"][f]['name']))
    date.append(str(rock['launches'][f]['windowend']))
    o_name.append(str(rock['launches'][f]['lsp']['name']))
    lat.append(float(rock["launches"][f]["location"]["pads"][0]["latitude"]))
    lng.append(float(rock["launches"][f]["location"]["pads"][0]["longitude"]))
    if str(rock["launches"][f]['failreason']) != "None":
        reason_f.append(str(rock["launches"][f]['failreason']))
        sf.append("F")
    else:
        sf.append("S")
        reason_f.append("Na")
        
rocksf = {"rockets name": name, "organization name": o_name, "date": date,"lat": lat, "lng": lng,
          "S/F": sf, "reason for failure": reason_f}
rocksf = pd.DataFrame(rocksf)
rocksf
```

```python
rocksf.to_csv("falcon.csv")
```
